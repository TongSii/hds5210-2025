# Value Counts on an already unique column?

We should get a series of counts where every count has a value of 1

```python
pop['NAME'].value_counts()
```

```python
(pop['NAME'].value_counts() == 1).all()
```

```python
(pop['DIVISION'].value_counts() == 1).any()
```
