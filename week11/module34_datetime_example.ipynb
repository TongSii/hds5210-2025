```python
import pandas as pd
import datetime
```

```python
ndc = pd.read_csv('product.txt',delimiter='\t',dtype=str)
```

```python
ndc.head()
```

```python
cert_through = ndc['LISTING_RECORD_CERTIFIED_THROUGH'].astype(str)
```

```python
end_date = pd.to_datetime(cert_through.str[0:8], format='%Y%m%d')
```

```python
end_date.head()
```

```python
ndc['end'] = end_date
```

```python
market_start = ndc['STARTMARKETINGDATE'].astype(str)
```

```python
ndc['start'] = pd.to_datetime(market_start.str[0:8], format='%Y%m%d')
```

```python
ndc.head(100)
```

# Date Differences


```python
ndc['duration']=ndc['end'] - ndc['start']
```

```python
ndc['duration'].count()
```

```python
ndc['duration'].isna().sum()
```

```python
ndc.head()
```

```python
import numpy as np
ndc['years'] = (
    ndc['duration'].fillna(pd.Timedelta(0)) / np.timedelta64(1, 'D') / 365.25
).astype(int)
```

```python
ndc.head()
```

```python
ndc.shape
```

```python
ndc['age rank']=ndc.years.rank()
```

```python
ndc[ndc.start > '1900-12-31'].sort_values('age rank', ascending=False).head()
```

```python
ndc.head()
```

```python
ndc[ndc.start > '1900-12-31'].sort_values('age rank',ascending=False).head()
```
