```python
#Query metadata
df = con.cursor().execute('''
    select current_account(),
           current_timestamp(),
           current_database(),
           current_schema(),
           current_warehouse();
''').fetch_pandas_all()

df
```

```python
import pandas as pd

df = pd.DataFrame({
    'LAST_NAME': ['Boal', 'Boal', 'Lester'],
    'FIRST_NAME': ['Paul', 'James', 'Sarahlynn']
})
```

```python
from snowflake.connector.pandas_tools import write_pandas

con.cursor().execute(
    'create table if not exists NAMES (LAST_NAME varchar, FIRST_NAME varchar)'
)
write_pandas(con, df, 'NAMES')
```

```python
try:
    con.close()
except:
    pass

con = None
```
