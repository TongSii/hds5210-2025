```python
def bmi(weight, height):
    """
    (float, float) -> float
    This function calculates BMI using the naive formula of
    the ratio between the weight of the individual and the
    square of the height of the individual.

    This version expects the weight and height to be in
    pounds and inches respectively.

    If the height or weight are less than or equal to zero
    the code will return None.

    >>> bmi(200, 72)
    27.12191358024691

    >>> bmi(120, 60)
    23.433333333333334
    """
    if height > 0 and weight > 0:
        return weight / height / height * 703
    else:
        return None
```

```python
from datetime import date

my_encounters = [
    { "id": "E1234", "admit": date(2019, 1, 3), "diagnosis": "COPD", "discharge": date(2019, 1, 8) },
    { "id": "E8342", "admit": date(2019, 1, 5), "diagnosis": "Hypertension", "discharge": date(2019, 1, 9) },
    { "id": "E9231", "admit": date(2019, 1, 12), "diagnosis": "Anxiety", "discharge": date(2019, 1, 13) }
]
```

```python

def los(admit, discharge):
    """
    (date, date) -> int
    Compute the number of whole days between the admit and discharge dates

    >>> los(date(2019,1,1), date(2019,1,6))
    5
    """
    return (discharge - admit).days
```

```python

def average_los(encounters):
    """
    (list) -> float
    Returns the average of the LOS for the encounters in the provided list.
    The encounters must be a list of dictionaries that each contain at least an admit date
    and a discharge date.
    """
    total_los = 0
    total_encounters = len(encounters)

    if total_encounters == 0:
        return 0

    for enc in encounters:
        total_los += los(enc['admit'], enc['discharge'])

    return total_los / total_encounters
```

```python
average_los(my_encounters)
```
